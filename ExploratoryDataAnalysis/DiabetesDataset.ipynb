```python
from sklearn.datasets import load_diabetes

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
diabetes_data = load_diabetes()
diabetes_data
```

```python
print(diabetes_data.DESCR)
```

```python
diabetes_data['data']
```

```python
diabetes_data['feature_names']
```

```python
diabetes_data['target']
```

```python
diabetes_df = pd.DataFrame(
                data=np.c_[diabetes_data['data'], diabetes_data['target']],
                columns=[diabetes_data['feature_names'] + ['target']])

diabetes_df
```

```python
diabetes_df.iloc[:,:-1].describe()
```

```python
rows = 5
cols = 2

fig, axs = plt.subplots(5,2, figsize=(20, 35))
for i in range(rows):
    for j in range(cols):
        axs[i][j].scatter(diabetes_df.iloc[:,i], diabetes_data['target'])
        axs[i][j].set_xlabel(diabetes_data['feature_names'][i*cols + j])
        axs[i][j].set_ylabel("Disease Progression")
plt.show()
```
