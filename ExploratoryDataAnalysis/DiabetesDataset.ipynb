# Exploratory Data Analysis On Diabetes Dataset

```python
from sklearn.datasets import load_diabetes

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

I will be performing exploratory data analysis (EDA) on [Scikit-Learn's Diabetes toy dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html#sklearn.datasets.load_diabetes).

EDA steps:

1. Load data
2. Examine loaded data structure
3. Extract description (if available)
4. Extract data and target values
5. Check for missing values
6. Obtain summary statistics

## 1. Load Data

```python
data = load_diabetes()
```

## 2. Exmaine Data Structure

```python
data
```

## 3. Extract Description

```python
print(data.DESCR)
```

## 4. Extract Data And Target Values

```python
# View the data
data['data']
```

```python
# View Feature names of the data
data['feature_names']
```

```python
# View the target values
data['target']
```

```python
# Create a data frame
df = pd.DataFrame(
                data=np.c_[data['data'], data['target']],
                columns=[data['feature_names'] + ['target']])

df
```

This data has been normalized. Notice how sex is not 0 and 1. And, it is mentioned in the description of the data (if you missed that part in the description).

## 5. Check For Null Values

There shouldn't be any null values in this toy dataset, but we will check anyways.

```python
# Check if there are any values whatsoever in the data frame
df.isnull().any().any()
```

## 6. Obtain Summary Statistics

```python
# Leave out the labels column when obtaining summary statistics of data
df.iloc[:,:-1].describe()
```

## 7. Visualizations

### 1. Histograms

Let's see the spread of the data

```python
import seaborn as sns

data_df = df.iloc[:,:-1]

# Define the number of rows based on the number of columns in the DataFrame
n_cols = 3
n_rows = int(np.ceil(df.shape[1] / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))

# Flatten the axes array, so we can easily iterate over it
axes = axes.ravel()

for i, column in enumerate(data_df.columns):
    sns.histplot(df[column], bins=50, kde=True, ax=axes[i])
    axes[i].set_title(column)
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Frequency")

# Remove any remaining empty subplots
for j in range(i+1, n_rows*n_cols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Observations** (note: feature descriptions are above in the notebook):

1. **age**: Data is fairly uniform suggesting there are roughly equal number of samples across different age groups in the dataset. Appears there may be more samples from older age groups though.

2. **sex**: A bimodal distribution that is normalized. This is expected.

3. **bmi**: Distribution is somewhat Gaussian (bell-shaped). Most samples bmi is around the mean.

4. **bp**: Distribution is somewhat Gaussian. Most samples bp is around the mean.

5. **s1**: Distribution is approximately Gaussian.

6. **s2**: Distribution is approximately Guassian.

7. **s3**: Slightly right skewed distribution. Indicates more samples with lower HDL.

8. **s4**: Slightly right skewed distribution. Indicates more samples with lower TCH.

9. **s5**: Distribuution is approximately Guassian.

10. **s6**: Distribution is approximately Guassian.

### 2. Box Plots

```python
# Plotting the boxplots for each feature
plt.figure(figsize=(16, 8))
df.drop(columns="target").boxplot()
plt.title("Boxplots of Features in the Diabetes Dataset")
plt.ylabel("Normalized Values")
plt.xticks(rotation=45)
plt.show()
```

### 3. Scatter Plots

```python
# Set up the figure and axis
fig, axs = plt.subplots(5, 2, figsize=(14, 18))
axs = axs.ravel()

# Plot scatter plots for each feature
for i in range(10):
    axs[i].scatter(df.iloc[:, i], df['target'], alpha=0.5)
    axs[i].set_title(f"Scatter plot of {data.feature_names[i]}")
    axs[i].set_xlabel(data.feature_names[i])
    axs[i].set_ylabel("Disease Progression")

plt.tight_layout()
plt.show()
```

Observations:

2. **Sex**: Cannot infer any type of correlation
3. **bmi**: Positive correlation
4. **bp**: Weak - Moderate postive correlation
