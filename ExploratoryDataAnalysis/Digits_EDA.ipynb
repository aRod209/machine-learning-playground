# Exploratory Data Analysis on Digits Dataset

Notebook by Anthony Rodriguez

This notebook does exploratory data analysis (EDA) on binary digits. This EDA will be used in future notebooks to classify the digits. The data is a [toy dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html#sklearn.datasets.load_digits) in SKLearn, but the dataset is a test/partial dataset originally from an [optical recognition of handwritten digits dataset](https://archive.ics.uci.edu/ml/datasets/Optical+Recognition+of+Handwritten+Digits). The images being used in this EDA are 8X8 images.

```python
from sklearn.datasets import load_digits

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

## Load Data

```python
digits = load_digits()
digits
```

## Data description

```python
print(digits.DESCR)
```

```python
digits.feature_names
```

```python
digits.target_names
```

```python
print(f'Number of elements in data: {digits.data.size}')
print(f'Shape of data: {digits.data.shape}')
print('Data:\n')
digits.data
```

```python
print(f'target:\n\n{digits.target}')
```

```python
df = pd.DataFrame(
            data=np.c_[digits.data, digits.target],
            columns=digits.feature_names + ['target_names'])
df
```

## Summary Statistics

```python
# Allow to see all columns
pd.set_option('display.max_columns', None)

# Display summary statistics
df.iloc[:, :-1].describe()
```

## Visualize Digits

Let's see what some of these images look like in their 8X8 form.

```python
plt.gray()

for i in range(10):
    plt.matshow(digits.images[i])

plt.show()
```

## Pixel Intensity Analysis

On average, let's see which digits are distinguishable by doing average pixel intensity analysis.

```python
target_pixel_mean_df = df.groupby('target_names').mean()
target_pixel_mean_df
```

```python
plt.gray()

for i in range(10):
    plt.matshow(np.resize(target_pixel_mean_df.iloc[i].to_numpy(), (8,8)))

plt.show()
```

On average, some numbers are distinguishable, such as 0 and 3. Some numbers not so much, such as 7 and 8.

## Visualize Clusters

Let's use t-Distributed Stochastic Neighbor Embedding (t-SNE) to visualize any clustering of the digits using the first two principal components.

```python
from sklearn.manifold import TSNE

# Perform t-SNE to reduce data to two dimensions
tsne = TSNE(n_components=2, random_state=42)
digits_tsne = tsne.fit_transform(digits.data)
```

```python
# Plotting the t-SNE reduced data
plt.figure(figsize=(10, 8))
scatter = plt.scatter(digits_tsne[:, 0], digits_tsne[:, 1], c=digits.target, cmap='tab10', alpha=0.6)
plt.title('2D t-SNE of the Digits Dataset')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.colorbar(scatter, label='Digit Label')
plt.show()
```

**Observations:**
- There seems to be 10 distinct clusters that are about the same size.
- Digit 1 has a small outlying cluster that is far from it's general cluster.
- The 3 digit has some outliers that is away from it's general cluster.
- The same can be said for digit 9.
